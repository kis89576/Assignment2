```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
df = pd.read_csv('DATA.csv', index_col=False)
#집계일자,집계시,영업소코드,입출구구분코드,TCS차종구분코드,TCS하이패스구분코드,교통량,최종변경시각 순서
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
print(df.head())
```

```python
df = df.drop(['FINAL_CHANGE_TIME'],axis=1)
print(df.head())
```

```python
df_traffic_per_day = df.groupby('COL_DATE')['TRAFFIC'].mean()
traffic_mean = df_traffic_per_day.mean()
df_traffic_per_day.plot(kind="bar")
plt.xlabel('DATE', size=15)
plt.ylabel('TRAFFIC',size=15)
plt.suptitle('TRAFFIC Per Date')
print(traffic_mean)
```

```python
df1 = df[df['HIPASS'] == 1]
df1 = df1[df1['GATE_CODE'] != 0]
df_hipass = df1.groupby('GATE_CODE')['TRAFFIC'].mean().sort_values(ascending=False)
df_hipass.head().plot(kind = 'bar')
plt.xlabel('GATE', size=15)
plt.ylabel('TRAFFIC', size=15)
plt.suptitle('CAR TRAFFIC HAS HIPASS')
```

```python
df2 = df[df['CARNUM'] == 1]#8종분류 의거 - 일반차량
df2 = df2[df2['GATE_CODE'] != 0]
df3 = df[df['CARNUM'] != 1] #8종분류 의거 - 화물차량(대형)
df3 = df3[df3['GATE_CODE'] != 0]
df7 = df2.groupby('COL_DATE')['TRAFFIC'].mean().sort_values(ascending=False)
df7.plot(kind='bar')
plt.xlabel('DATE')
plt.ylabel('TRAFFIC')
plt.suptitle('TRAFFIC per Date(General car)')
plt.show()
df8 = df3.groupby('COL_DATE')['TRAFFIC'].mean().sort_values(ascending=False)
df8.plot(kind='bar')
plt.xlabel('DATE')
plt.ylabel('TRAFFIC')
plt.suptitle('TRAFFIC per Date(Huge car)')
plt.show()
df2_per_gate = df2.groupby('GATE_CODE')['TRAFFIC'].mean().sort_values(ascending=False)
df2_per_gate.head().plot(kind='bar')
plt.xlabel('GATE')
plt.ylabel('TRAFFIC')
plt.suptitle('TRAFFIC Per gate(top5) - General Car')
plt.show()
df3_per_gate = df3.groupby('GATE_CODE')['TRAFFIC'].mean().sort_values(ascending=False)
df3_per_gate.head().plot(kind='bar')
plt.xlabel('GATE')
plt.ylabel('TRAFFIC')
plt.suptitle('TRAFFIC per gate(top5) - Huge Car')
plt.show()
df4 = df[df['COL_TIME'] <= 7]
df5 = df[df['COL_TIME'] >= 22]
df_nd = pd.concat([df4,df5],ignore_index=True) #심야시간 주행차량
df_nd_1 = df_nd[df_nd['CARNUM'] == 1]
df_nd_2 = df_nd[df_nd['CARNUM'] != 1]
df_night_drive1 = df_nd_1.groupby('COL_DATE')['TRAFFIC'].mean()
df_col1 = df7 / df_night_drive1
df_col1.plot(kind = 'bar')
plt.xlabel('COL_DATE')
plt.ylabel('PERCENT')
plt.suptitle("주간 주행 / 야간 주행 (일반 차량)")
plt.show()
#print(df_col1)
df_night_drive2 = df_nd_2.groupby('COL_DATE')['TRAFFIC'].mean()
df_col2 = df8 / df_night_drive2
df_col2.plot(kind='bar')
plt.xlabel('COL_DATE')
plt.ylabel('PERCENT')
plt.suptitle("주간 주행 / 야간 주행 (대형 차량)")
plt.show()
#print(df_col2)
```
